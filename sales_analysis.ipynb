```python
import matplotlib.pyplot as plt
sales_by_category = df.groupby("category")["sales"].sum()
sales_by_category.plot(kind="bar")
plt.title("sales by category")
plt.xlabel("category")
plt.ylabel("sales")
plt.show()
```

```python
sales_by_region = df.groupby("region")["sales"].sum()
sales_by_region.plot(kind="bar")
plt.title("sales by region")
plt.xlabel("region")
plt.ylabel("sales")
plt.show()
```

1.Technology category generates the highest sales, indicating strong demand for tech products.

2.West/East region performs better compared to other regions.

3.Sales are uneven across regions, suggesting opportunities for targeted marketing.

4.Created bar charts using Matplotlib for visualization.
